```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data into a Pandas DataFrame
data = pd.read_excel('pokemon_data.xlsx')
```

```python
# Convert spawn time to seconds for easier analysis
data['spawn_time_seconds'] = pd.to_datetime(data['spawn_time']).dt.minute * 60 + pd.to_datetime(data['spawn_time']).dt.second
```

```python

# Get all Pokemons whose spawn rate is less than 5%
spawn_rate_less_than_5 = data[data['spawn_chance'] < 0.05]
print("Pokemons with spawn rate less than 5%:")
print(spawn_rate_less_than_5)

# Plot spawn rate distribution
plt.hist(data['spawn_chance'], bins=20)
plt.xlabel('Spawn Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Spawn Rate')
plt.show()
```

```python


# Get all Pokemons that have less than 4 weakness

temp = [index for index in data['weakness'].index if len(data['weakness'][index].split(','))<4]
less_than_4_weakness = data.iloc[temp]
print("Pokemons with less than 4 weaknesses:")
print(less_than_4_weakness)

# Plot weakness distribution
plt.hist(data['weakness'], bins=range(0, 10))
plt.xlabel('Number of weaknesses')
plt.ylabel('Frequency')
plt.title('Distribution of weaknesses')
plt.xticks(range(0, 10), rotation=45)
plt.show()
```

```python

# Get all Pokemons that have no multipliers at all
no_multipliers = data[data['multipliers'] == 'NaN']
print("Pokemons with no multipliers:")
print(no_multipliers)
```

```python
# Get all Pokemons that do not have more than 2 type

temp = [index for index in data['type'].index if len(data['type'][index].split(','))<3]
less_than_2_type = data.iloc[temp]
print("Pokemons with less than 3 type:")
print(less_than_2_type)

# Plot type distribution
plt.hist(data['type'], bins=range(0, 10))
plt.xlabel('Number of Evolutions')
plt.ylabel('Frequency')
plt.title('Distribution of Evolutions')
plt.xticks(range(0, 10), rotation=90)
plt.show()
```

```python
# Get all Pokemons whose spawn time is less than 300 seconds
spawn_time_less_than_300 = data[data['spawn_time_seconds'] < 300]
print("Pokemons with spawn time less than 300 seconds:")
print(spawn_time_less_than_300)

# Plot spawn time distribution
plt.hist(data['spawn_time_seconds'], bins=20)
plt.xlabel('Spawn Time (seconds)')
plt.ylabel('Frequency')
plt.title('Distribution of Spawn Time')
plt.show()
```
