```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the data into a Pandas DataFrame
data = pd.read_csv('meteorite_data.csv')
```

```python
# Convert the year column to datetime for easier analysis
data['year'] = pd.to_datetime(data['year'], errors='coerce').dt.year
```

```python
# Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = data[(data['reclat'].notnull()) & (data['year'] < 2000)]
print("Earth meteorites that fell before the year 2000:")
print(earth_meteorites_before_2000)
```

```python
# Get all the Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_coordinates_before_1970 = data[(data['reclat'].notnull()) & (data['year'] < 1970)][['reclat', 'reclong']]
print("Earth meteorites' coordinates that fell before the year 1970:")
print(earth_meteorites_coordinates_before_1970)
```

```python
# Plot the distribution of meteorite falls over time
plt.hist(data['year'].dropna(), bins=range(1800, 2025, 25))
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Distribution of Meteorite Falls Over Time')
plt.show()
```

```python
# Assuming mass is in kg, get all Earth meteorites with mass greater than 10000 kg
mass_greater_than_10000kg = data[(data['reclat'].notnull()) & (data['mass'].notnull()) & (data['mass'] > 10000)][['name', 'mass']]
mass_greater_than_10000kg['mass (tonnes)'] = mass_greater_than_10000kg['mass'] / 1000  # Convert mass to kg
print("Earth meteorites with mass greater than 10000 kg:")
print(mass_greater_than_10000kg)
```

```python
# Plot the distribution of meteorite masses
plt.hist(data['mass'].dropna(), bins=range(0, 200000, 10000))
plt.xlabel('mass')
plt.ylabel('Frequency')
plt.title('Distribution of Meteorite Masses')
plt.show()
```
